## slicing on array

```python
a = np.arange(1,10) 
a
```

```python
a [0:3] # from index 0 to 3
```

```python
a [-1:5:-1] # first -1 is for start and 5 is for end step and last -1 is reverse step
```

```python
a [2:6] # from index 2 to 6
```

```python
a [0:] # from 0 index to end
```

```python
a [0::1] # from 0 to end in positive way
```

```python
a [0::2] # from 0 to end and 2 is step
```

```python
a [::2] # 0 to end in 2 step
```

```python
a [::-1] # in reverse 0 to end
```
