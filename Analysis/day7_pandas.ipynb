```python
df = pd.DataFrame({
   "Team"    : ['A' , 'A' , 'B' , 'B' , 'C' , 'C'] ,
    "Player" : ['P1' , 'P2' , 'P3' , 'P4' , 'P5' , 'P6'] , 
    "Score"  : [15 , 18 , 10 , 25 , 20 , 23]
})
```

```python
df
```

```python
def score_diff(x):
    return x.max() - x.min()

ans = df.groupby("Team")["Score"].apply(score_diff)  # apply se udf function auto call ho jata hai
#ans = df.groupby("Team")["Score"]
ans
```

```python
# filter groups with team total score > 35

total_score = df.groupby("Team")["Score"].sum()
#print(total_score)

high_score_teams = total_score[total_score>35].index 
#print(high_score_teams)

filtered_df = df[df["Team"].isin(high_score_teams)]
filtered_df
```

### Transform function

```python
df
```

```python
df ["New_score"] = df["Score"].transform(lambda x: x+5) # add 5 in score and add 1 new columns of new score
df
```

```python
def disp(n):
    return n + 10

df["Diff_score"] = df["Score"].transform(disp) # add 10 in score and add 1 new columns of Diff score by UDF
df
```

```python
df
```

```python
df.drop(columns = ["New_score" , "Diff_score"] , inplace = True) # remove added 2 columns
```

```python
df
```

```python
df["Avg_score"] = df.groupby(by = "Team")["Score"].transform("mean") # agg function - mean
```
