## join & Merge data

### type if merge

- inner (matching)
- outer (full)
- left  (left side all and from in right side that part which is available in left round  )
- right (right side all and from in left side that part which is available in right round )

```python
df1 = pd.DataFrame({
    "Id" : [1 , 2 , 3],
    "Name" : ["charlie" , "Bob" , "John"],
})
```

```python
df1
```

```python

df2 =pd.DataFrame({
   "Id" : [ 2, 3 , 4] ,
    "Age" :[  20 , 25 , 60]
})
                  
```

```python
df2
```

```python
pd.merge(df1 , df2 , on="Id" , how = "left")
```

```python
pd.merge(df1 , df2 , on="Id" , how = "right")
```

```python
pd.merge(df1 , df2 , on="Id" , how = "inner")
```

```python
pd.merge(df1 , df2 , on="Id" , how = "outer")
```

```python
pd.merge(df1 , df2)  # by dafult merge by inner
 
```

```python
df1.merge(df2)
```
