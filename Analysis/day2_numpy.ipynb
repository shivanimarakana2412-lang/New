## searching

```python
a1 = np.array([12 , 35 , 58 , 90])
a1
```

```python
print(np.where(a1 == 12)) # give index of value from array 
print(np.where(a1 == 35))
print(np.where(a1 == 58))
print(np.where(a1 == 90))
```

```python
a2 = np.array([1 , 5 , 6 , 9 , 7 , 3 , 5 , 5])
a2
```

```python
np.where(a2 == 5) # give indexes of value(5) from array 
```

```python
np.where(a2 > 5) # give index of big number than 5
```
