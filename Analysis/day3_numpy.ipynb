## Broadcasting of Array

Broadcasting is a powerful meachnism is NumPy that allows array operations between arrays of different shapes without copying data.

In simple terms : Brodcasting allows you to automatically expand smaller arrays to match the shape of larger arrays during operatins like addition , multiplication , etc...

### Types of brodcasting

- scalar brodcasting
- Vactor brodcasting
- Matrix brodcasting

# scalar brodcasting

```python
# one element array
```

```python
arr
```

```python
s1  = [3]
s2 = 5 # this value add in every vactor
s3 = np.array([6])
```

```python
print(arr + s1)
```

```python
arr + s2
```

```python
arr + s3
```

# vactor brodcasting

- 1. row vactor
- 2. column vactor

# row vactor

```python
a1 = np.array([[1 , 2 , 3] , [4 , 5 , 6]])  # shape (2,3)
a1
```

```python
a2 = np.array([[10 , 20 , 30]]) # shape (3,)
a2
```

```python
print(a1 + a2)
# a1.shape = (2,3)
# a2.shape = (3, )
```

# column vactor

```python
a1
```

```python
a2 = np.array([[10] , [20]])
a2
```

```python
a2 = np.array([10,20]).reshape(-1 , 1)
a2
```

```python
print(a1 + a2)
```

```python
a2.shape = (2 , 1)
a2
```

## Matrix Brodcasting

#### Brodcasting happens between matrix of different shapes if they satisfy brodcasting rules.

```python
a1 = np.array([[1],[2]]) # 1,1,1
                         # 2,2,2
                         
a2 = np.array([[10 , 20 ,30]]) #10 20 30
                               #10 20 30
```

```python
print(a1)
print()
print(a2)
print()
```

```python
a3 = a1 + a2
print(a3)
```

```python
# a1 --> (2 , 1) -->expened across columns
# a2 --> (1 , 3) -->expened across raws
# result shape  --> (2 , 3)
```
