# Handling Categorical Data

```python
df["Color"] = df["Color"].astype("category")
```

```python
df = pd.DataFrame({
   'Size' : ['M' , 'S' , 'L' , 'M'] ,
   'Iteam' : ['shirt' , 'pants' , 'shirt' , 'cap']
    
})
```
