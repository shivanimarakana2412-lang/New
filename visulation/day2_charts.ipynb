```python
import matplotlib.pyplot as plt
```

# Stack charts

```python
days = ['Sun' , ' Mon' , 'Tue' , 'Wed' , 'Thu']
product_A = [5, 7, 6, 4, 5]
product_B = [3, 5, 2, 6, 4]
product_C = [2, 3, 4, 3, 2]
```

```python
# plt.bar(days, product_A, label='Product A')
# plt.bar(days, product_B, label='Product B')
# plt.bar(days, product_C, label='Product C')

plt.plot(days, product_A, label='Product A')
plt.plot(days, product_B, label='Product B')
plt.plot(days, product_C, label='Product C')

plt.legend()
plt.show()
```

```python
plt.bar(days, product_A, label='Product A' , width = 0.5)
plt.bar(days, product_B, label='Product B' , width = 0.4)
plt.bar(days, product_C, label='Product C' , width = 0.3)


plt.legend()
plt.show()
```

```python
plt.barh(days, product_A, label='Product A')
plt.barh(days, product_B, label='Product B')
plt.barh(days, product_C, label='Product C')

plt.legend()
plt.show()
```

# Stacked Area Chart

```python
plt.stackplot(
    days , 
    product_A ,
    product_B ,
    product_C ,
    labels= ['Product A' , 'Product B' , 'Product C' ]   
)

plt.legend()
plt.grid(axis = 'y' , alpha = 0.4)
plt.show()
```

# Subplot

```python
# syntax
#------------
#fig , axs = plt.subplot
```

```python
x = [1 , 2 , 3 , 4 , 5]
y1 = [10 , 20 , 30 , 40 , 50 ]
y2 = [5 , 15 , 25 , 35 , 45]

fig,axs = plt.subplots(1 , 2  , figsize = (10 , 4))

axs[0].plot(x , y1 , color = 'pink')
axs[0].set_title("LINE PLOT")

axs[1].bar(x , y2 , color = 'lightblue')
axs[1].set_title("BAR CHART")

plt.tight_layout()
plt.show()
```

```python
# Multiple chart by data

x = [1 , 2 , 3 , 4 , 5]
y = [2 , 4 , 6 , 8 , 10]

fig,axs = plt.subplots(2 , 2  , figsize = (10 , 6))

axs[0 , 0].plot(x , y , color = 'pink')
axs[0 , 0].set_title("LINE PLOT 1")

axs[0 , 1].bar(x , y , color = 'lightblue')
axs[0 , 1].set_title("BAR CHART 1")


axs[1 , 0].scatter(x , y , color = 'lightgreen')
axs[1 , 0].set_title("SCATTER 1")

axs[1 , 1].barh(x , y , color = 'pink')
axs[1 , 1].set_title("HORIZONTAL BAR")

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

x = [1 , 2 , 3 , 4]
y1 = [1 , 4 , 2 , 3]
y2 = [2 , 3 , 4 , 1]

plt.plot(x , y1 , label = 'Line 1')
plt.plot(x , y2 , label = 'Line 2')

plt.title("Legend Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# plt.legend() # Show legend
plt.legend( loc = "upper left")         # position
plt.legend(fontsize = 12)               # FOnt size
plt.legend(title = "Lines")             # Title for legend box
plt.legend(frameon = True)              # For box frame
plt.show()
```

```python

x = [1 , 2 , 3 , 4]
y1 = [1 , 4 , 2 , 3]
y2 = [2 , 3 , 4 , 1]

plt.plot(x , y1 , label = 'Line 1')
plt.plot(x , y2 , label = 'Line 2')

plt.title("Legend Out side the Polt")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.legend(
     loc = "upper left" , 
     fontsize = 'medium',
     title = "Legend Title" ,
     frameon = True ,
     shadow = True ,
     title_fontsize = 'large' ,
     facecolor = 'lightyellow' ,
     edgecolor = 'black'
)

plt.tight_layout()
plt.show()
```

```python

x = [1 , 2 , 3 , 4]
y1 = [1 , 4 , 2 , 3]
y2 = [2 , 3 , 4 , 1]

plt.plot(x , y1 , label = 'Line 1')
plt.plot(x , y2 , label = 'Line 2')

plt.title("Legend Out side the Polt")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.legend(loc = 'center left' , bbox_to_anchor = (1 , 0.5))

plt.show()
```

```python
# multiple legends

line1, = plt.plot(x , y1 , label = 'Line A')
line2, = plt.plot(x , y2 , label = 'Line B')

first_legend = plt.legend(handles = [line1] , loc = 'upper left')
plt.gca().add_artist(first_legend)                      # add the first legend manually
plt.legend(handles = [line2] , loc = "lower right")     # second legend
```
