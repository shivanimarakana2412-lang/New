# Bar charts

- Bar charts is use for comperasion.

```python

x = ['Math', 'Science', 'Eng' , 'Ss' , 'Acc' , 'Stat' , 'Eco']
y = [85, 90, 70 , 89 , 76 , 56 , 65]

plt.bar(x, y , color = "pink")
plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
plt.bar(x, y, width = 0.5 ) # width must be in range 0.0 to 1.0
plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
plt.bar(x, y, width = 0.5, color = "pink")
plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
plt.bar(x, y, width = 0.5, color= ['pink' , 'lightblue' ])
plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
plt.bar(x, y, width = 0.5, color= ['pink' , 'lightblue' ] , alpha =0.5) # alpha (opacity) must be in range 0.0 to 1.0
plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python

plt.figure(figsize = (4 , 2)) #  figsize is in inches
plt.bar(x, y,align = 'center', width = 0.5, color= ['pink' , 'lightblue' ], alpha =0.5)
plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
plt.bar(x, y,align = 'center', width = 0.5,color= ['pink' , 'lightblue' ] , edgecolor = 'black'  ,linewidth =2)
plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
plt.bar(x , y , align = 'center', width = 0.5, color= ['pink' , 'lightblue' ] , edgecolor = 'blue'  ,linewidth =2)

plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')

# # # opetional thing we can add

plt.text(0 , 85+1 , str(85) , ha = "center" , fontweight = "bold")
plt.text(1 , 90+1 , str(90) , ha = "center" , fontweight = "bold")
plt.text(2 , 70+1 , str(70) , ha = "center" , fontweight = "bold")
plt.text(3 , 89+1 , str(89) , ha = "center" , fontweight = "bold")
plt.text(4 , 76+1 , str(76) , ha = "center" , fontweight = "bold")
plt.text(5 , 56+1 , str(56) , ha = "center" , fontweight = "bold")
plt.text(6 , 65+1 , str(65) , ha = "center" , fontweight = "bold")

 

# for i , v in enumerate(y):
#    plt.text(i , v+1 , str(v) , ha = "center" , fontweight = "bold")

plt.tight_layout()
plt.show()

```

# Line chart

- Show growth

```python
months = ['Jan' , 'Feb' , 'March' , 'April' , 'May' , 'June' , 
          'July' , 'Aug' , 'Sep' , 'Oct' , 'Nov' , 'Dec']

revenue = [1000 , 2000 , 3000 , 4000 , 5000 , 6000 ,
           7000 , 8000 , 9000 , 10000 , 11000 , 12000]
```

```python
plt.plot(months , revenue , marker = 'o' , linestyle = '-' , color = 'pink' )
# plt.plot(months , revenue , marker = '+' , linestyle = '-' , color = 'pink' )
# plt.plot(months , revenue , marker = '*' , linestyle = '-' , color = 'pink' )

plt.title('revenue by months')
plt.xlabel('months')
plt.ylabel('revenue')
plt.tight_layout()
plt.show()
```

```python
plt.title('revenue by months')
plt.xlabel('months')
plt.ylabel('revenue')
plt.tight_layout()
plt.show()
```

```python
plt.plot(months , revenue , marker = 'o' , linestyle = '-' , color = 'black' , markerfacecolor = 'red' , markersize = '8' )

plt.title('revenue by months')
plt.xlabel('months')
plt.ylabel('revenue')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize= (12 , 8))
plt.plot(months , revenue , marker = 'o' , linestyle = '-' , color = 'black' , markerfacecolor = 'red' , markersize = '8' )

plt.title('revenue by months')
plt.xlabel('months')
plt.ylabel('revenue')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize= (12 , 8))
plt.plot(months , revenue , marker = 'o' , linestyle = '-' , color = 'black' , markerfacecolor = 'pink' , markersize = '8' )

plt.title('revenue by months')
plt.xlabel('months')
plt.ylabel('revenue')

for i , val in enumerate(revenue):
    plt.text(i , val + 250 , str(val) , ha = 'center' , fontweight = 'bold')

plt.grid(True, linestyle='--', alpha=0.5)
plt.savefig('line_chart.png')# we only use png

plt.tight_layout()
plt.show()
```

# Scatter chart

- For show relatationship between two numerical values

```python
ad_spend = [5, 6, 7, 6, 8, 7.5, 9, 10, 9.5, 11, 10.5, 12]
sales = [40, 45, 50, 46, 54, 53, 58, 62, 59, 65, 64, 70]
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
plt.scatter(ad_spend , sales)
plt.show()
```

```python
plt.figure(figsize = (10 , 8))
plt.scatter(ad_spend , sales , color = 'pink' , edgecolors = 'black' , s = 100) # s is size of marker
plt.show()
```

```python
plt.figure(figsize = (10 , 5))
plt.scatter(ad_spend , sales ,color = 'pink' , edgecolors = 'black' , s = 50 ,)

plt.xlabel('ad_spend')
plt.ylabel('sales')

for i , month in enumerate(months):
    plt.text(ad_spend[i] , sales[i] + 1.5 , months[i] , ha = 'center' , fontweight = 'bold')

plt.show()
```

# Pie chart


 
- A Pie Chart shows percentage or proportion of different categories as slices of a circle.

```python
# Data
labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 45, 25]   # Must add up to 100%

# Create Pie Chart
plt.pie(sizes, labels=labels)
plt.show()
```

```python

labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 30, 20, 25]

plt.pie(sizes, labels=labels)
plt.title('Fruit Sales Distribution')
plt.show()
```

```python

labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [20, 40, 10, 30]
colors = ['red', 'yellow', 'pink', 'brown']

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%.1f%%',
    startangle=90
)
plt.title('Fruit Market Share')
plt.show()
```

```python
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [15, 50, 10, 25]
explode = (0, 0.1, 0, 0)  # Only Banana pops out!

plt.pie(
    sizes,
    labels=labels,
    explode=explode,
    autopct='%1.1f%%',
    shadow=True,
    startangle=140
)
plt.title('Fruit Popularity')
plt.show()
```

```python
products = ['Laptop', 'Tablet', 'Smartphone', 'Accessories']
sales = [400, 150, 300, 150]

# Create Pie
plt.pie(
    sales,
    labels=products,
    autopct='%1.0f%%',
    startangle=140,
    colors=['#4CAF50', '#FFC107', '#2196F3', '#9C27B0']
)

plt.title('Product Sales Distribution')
plt.show()
```

```python
products = ['Laptop', 'Tablet', 'Smartphone', 'Accessories']
sales = [400, 150, 300, 150]

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%.1f%%',
    startangle=90,
    explode=explode,
    shadow=True
)
plt.title('Product Sales Distribution')
plt.show()
```

# Histogram - Hist

- It is used to show Distribute numerical data

```python
# sample data : Math exam scores

scores = [45 , 76 , 25 , 46 , 19 , 38 , 92 , 12 , 78 , 96 ,
          59 , 85 , 67 , 49 , 68 , 58 , 67 , 95 , 58 , 46 ,
          68 , 89 , 73 , 67 , 64 , 92 , 10 , 73 , 64 , 98 ]
```

```python
#plt.hist(scores)
plt.hist(scores , bins = 15 )
```

```python
plt.hist(scores , bins = [45 , 55 , 64 , 75 , 85 , 89 ] , edgecolor = 'black' )
```

```python
# Sample Data

math_scores =    [45, 78, 56, 89, 90, 67, 80, 56, 92, 45,
                  60, 75, 85, 55, 70, 90, 88, 72, 60, 65,
                  68, 77, 85, 95, 48, 52, 60, 75, 80, 82]

science_scores = [40, 70, 60, 85, 88, 66, 75, 50, 90, 42,
                  58, 72, 80, 50, 65, 88, 86, 70, 55, 60,
                  63, 74, 83, 91, 45, 51, 58, 73, 78, 79]
```

```python
plt.figure(figsize=(10, 6))

# Plot Math Scores
plt.hist(math_scores, bins=7, color='blue', alpha=0.5, label='Math', edgecolor='black')

# Plot Science Scores
plt.hist(science_scores, bins=7, color='green', alpha=0.5, label='Science', edgecolor='black')

# Title and Labels
plt.title('Score Distribution: Math vs Science', fontsize=16)
plt.xlabel('Score Range', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)

# Legend and Grid
plt.legend(loc = 'best') 
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
