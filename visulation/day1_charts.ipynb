# Bar charts

- Bar charts is use for comperasion.

```python

x = ['Math', 'Science', 'Eng' , 'Ss' , 'Acc' , 'Stat' , 'Eco']
y = [85, 90, 70 , 89 , 76 , 56 , 65]

plt.bar(x, y , color = "pink")
plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
plt.bar(x, y, width = 0.5 ) # width must be in range 0.0 to 1.0
plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
plt.bar(x, y, width = 0.5, color = "pink")
plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
plt.bar(x, y, width = 0.5, color= ['pink' , 'lightblue' ])
plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
plt.bar(x, y, width = 0.5, color= ['pink' , 'lightblue' ] , alpha =0.5) # alpha (opacity) must be in range 0.0 to 1.0
plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python

plt.figure(figsize = (4 , 2)) #  figsize is in inches
plt.bar(x, y,align = 'center', width = 0.5, color= ['pink' , 'lightblue' ], alpha =0.5)
plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
plt.bar(x, y,align = 'center', width = 0.5,color= ['pink' , 'lightblue' ] , edgecolor = 'black'  ,linewidth =2)
plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
plt.bar(x , y , align = 'center', width = 0.5, color= ['pink' , 'lightblue' ] , edgecolor = 'blue'  ,linewidth =2)

plt.title('Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')

# # # opetional thing we can add

plt.text(0 , 85+1 , str(85) , ha = "center" , fontweight = "bold")
plt.text(1 , 90+1 , str(90) , ha = "center" , fontweight = "bold")
plt.text(2 , 70+1 , str(70) , ha = "center" , fontweight = "bold")
plt.text(3 , 89+1 , str(89) , ha = "center" , fontweight = "bold")
plt.text(4 , 76+1 , str(76) , ha = "center" , fontweight = "bold")
plt.text(5 , 56+1 , str(56) , ha = "center" , fontweight = "bold")
plt.text(6 , 65+1 , str(65) , ha = "center" , fontweight = "bold")

 

# for i , v in enumerate(y):
#    plt.text(i , v+1 , str(v) , ha = "center" , fontweight = "bold")

plt.tight_layout()
plt.show()

```

```python
plt.bar(x , y , align = 'center', width = 0.5, color= ['pink' , 'lightblue' ] , edgecolor = 'blue'  ,linewidth =2 , alpha = 0.5)

for i , v in enumerate(y):
   plt.text(i , v+1 , str(v) , ha = "center" , fontweight = "bold")
   
plt.tight_layout()
plt.show()
```

# Line chart

- Show growth

```python
months = ['Jan' , 'Feb' , 'March' , 'April' , 'May' , 'June' , 
          'July' , 'Aug' , 'Sep' , 'Oct' , 'Nov' , 'Dec']

revenue = [1000 , 2000 , 3000 , 4000 , 5000 , 6000 ,
           7000 , 8000 , 9000 , 10000 , 11000 , 12000]
```

```python
plt.plot(months , revenue , marker = 'o' , linestyle = '-' , color = 'pink' )
# plt.plot(months , revenue , marker = '+' , linestyle = '-' , color = 'pink' )
# plt.plot(months , revenue , marker = '*' , linestyle = '-' , color = 'pink' )

plt.title('revenue by months')
plt.xlabel('months')
plt.ylabel('revenue')
plt.tight_layout()
plt.show()
```

```python
plt.title('revenue by months')
plt.xlabel('months')
plt.ylabel('revenue')
plt.tight_layout()
plt.show()
```

```python
plt.plot(months , revenue , marker = 'o' , linestyle = '-' , color = 'black' , markerfacecolor = 'red' , markersize = '8' )

plt.title('revenue by months')
plt.xlabel('months')
plt.ylabel('revenue')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize= (12 , 8))
plt.plot(months , revenue , marker = 'o' , linestyle = '-' , color = 'black' , markerfacecolor = 'red' , markersize = '8' )

plt.title('revenue by months')
plt.xlabel('months')
plt.ylabel('revenue')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize= (12 , 8))
plt.plot(months , revenue , marker = 'o' , linestyle = '-' , color = 'black' , markerfacecolor = 'pink' , markersize = '8' )

plt.title('revenue by months')
plt.xlabel('months')
plt.ylabel('revenue')

for i , val in enumerate(revenue):
    plt.text(i , val + 250 , str(val) , ha = 'center' , fontweight = 'bold')

plt.grid(True, linestyle='--', alpha=0.5)
plt.savefig('line_chart.png')# we only use png

plt.tight_layout()
plt.show()
```

# for save and add grid

- plt.grid(True, linestyle='--', alpha=0.5) # for grid means like graph
- plt.savefig('line_chart.png') # we only use png

# Scatter chart

- For show relatationship between two numerical values

```python
ad_spend = [5, 6, 7, 6, 8, 7.5, 9, 10, 9.5, 11, 10.5, 12]
sales = [40, 45, 50, 46, 54, 53, 58, 62, 59, 65, 64, 70]
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
plt.scatter(ad_spend , sales)
plt.show()
```

```python
plt.figure(figsize = (10 , 8))
plt.scatter(ad_spend , sales , color = 'pink' , edgecolors = 'black' , s = 100) # s is size of marker
plt.show()
```

```python
plt.figure(figsize = (10 , 5))
plt.scatter(ad_spend , sales ,color = 'pink' , edgecolors = 'black' , s = 50 ,)

plt.xlabel('ad_spend')
plt.ylabel('sales')

for i , month in enumerate(months):
    plt.text(ad_spend[i] , sales[i] + 1.5 , months[i] , ha = 'center' , fontweight = 'bold')

plt.show()
```
