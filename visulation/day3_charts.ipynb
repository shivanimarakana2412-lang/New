```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Bar plot

```python
# load data set

df = sns.load_dataset('titanic')

plt.figure(figsize=(8,5))
sns.barplot(x = "class" , y = "fare" , data = df , estimator = np.sum , errorbar = None )  
plt.xlabel('Passanger class')
plt.ylabel("Total fare")
plt.title("Total fare by class")
plt.show()
```

```python
# Real - world styled datase

df = pd.DataFrame({
    'Category' : ['Laptop' ,'Laptop' ,'Laptop' , 'Phone' ,'Phone' , 'Tablet' ,'Tablet' ,'Tablet'] ,
    'Rating' : [4.5 , 5.6 , 4.8 , 7.6 , 4.0 , 4.1 , 3.9 , 4.6]
})

# Bar plot 

sns.barplot(x = 'Category' , y = 'Rating' , data = df , errorbar = None , palette='pastel' , hue = 'Category')
plt.title("Average Customer rating by product category")
plt.xlabel("Avg Rating")
plt.ylabel("Product category") 
plt.show()
```

```python
df = pd.DataFrame({
    'Category' : ['Laptop' ,'Laptop' , 'Phone' ,'Phone' , 'Tablet' ,'Tablet' ] ,
    'Gender' : ['Male' , 'Female' , 'Male' , 'Female' , 'Male' , 'Female'] ,
    'Rating' : [4.5 , 5.6 , 4.8  , 4.0 , 4.1 , 4.6]
})

sns.barplot(x = 'Category' , y = 'Rating' , data = df , errorbar = None , palette='Set2' , hue = 'Gender')
plt.title("Avg Rating by product & Gender")
plt.show()

```

## Line plot - Fights Dataset

```python
df = sns.load_dataset('flights')

flights = sns.load_dataset('flights')
line_data = flights.groupby('year')['passengers'].sum().reset_index()

# line_data

plt.figure(figsize=(8,5))
sns.lineplot(data = line_data , x = 'year' , y = 'passengers' , marker = 'o')
plt.title("Total Passangers per year")
plt.show()
```

## Scatter plot with regression line

```python
tips = sns.load_dataset("tips")
plt.Figure(figsize = (8 , 5))
sns.scatterplot(data = tips , x = 'total_bill' , y = 'tip' , hue = 'time')
# sns.regplot(data = tips , x = 'total_bill' , y = 'tip' , scatter = True , color = 'pink') # regression line
plt.title('Total bill vs Tip with regression line')
plt.show()
```

## pie chart

```python
# pie and donate chart is not available in seaborn

flights = sns.load_dataset('flights')
pie_data = flights.groupby('year')['passengers'].sum()
plt.figure(figsize=(7,7))
plt.pie(pie_data, labels=pie_data.index, autopct='%1.1f%%')
plt.title("Passenger Distribution by Year")
plt.show()

# donate chart
plt.figure(figsize=(7,7))
plt.pie(pie_data, labels=pie_data.index, autopct='%1.1f%%', wedgeprops={'width':0.4})
plt.title("Passengers by Class")
plt.show()
```

```python
df = sns.load_dataset('titanic')

pie_data = df['class'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(pie_data, labels=pie_data.index, autopct='%1.1f%%')
plt.title("Passengers by Class")
plt.show()
```

## Hist plot

```python
df = sns.load_dataset('titanic')

plt.figure(figsize=(8,5))
sns.histplot(df["age"].dropna() , bins=30, kde=True) # kde is for line
plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

## Count plot

```python
plt.figure(figsize=(8,5))
sns.countplot(data = df , x = 'class' , hue = 'sex')  # y ki value count ho ke aayegi
plt.title('Passanger count by title')
plt.show()
```

## Heat map - correlation in titanic Dataset

```python
df = sns.load_dataset('tips')
corr = df.corr(numeric_only = True ) 

#corr - it ts built in function

plt.figure(figsize=(10 , 6))
# sns.heatmap(corr , annot=True , linewidths=0.5 , cmap = 'YlOrRd')
sns.heatmap(corr , annot=True , linewidths=0.5 , cmap = 'RdPu')

plt.title('correlation heamap - titanic')
plt.show()
```
